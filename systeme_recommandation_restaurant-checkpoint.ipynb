# Système de Recommandation - Restaurant

## Objectif
Créer un système intelligent de recommandation de plats utilisant du Machine Learning.

```python
import pandas as pd 
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt

print("Bibliotheques importees")
```

```python
#creation des plats

plats_data = {
    'id' : [1, 2, 3, 4, 5],
    'nom' : ['poulet DG', 'poisson braise', 'ndole', 'poulet roti', 'sauce gombo'],
    'categorie' : ['viande', 'poisson', 'legumes', 'viande', 'legumes'],
    'ingredients' : [
        'poulet platain legumes epices',
        'poisson tomate oignons citrons',
        'feuilles ndole crevettes huile',
        'poulet epice huie',
        'gombo viande poisson'
    ],
    'prix': [3500, 4000, 3000, 3200, 2500],
    'temps de preparation' : [45, 30, 40, 35, 120]
}

plats_df = pd.DataFrame(plats_data)
print("plats creees:")
plats_df
```

```python
#creation des cpmmandes

commandes_data = {
    'id_commande' : ['CMD001', 'CMD002', 'CMD003', 'CMD004', 'CMD005', 'CMD006'],
    'id_plat' : [1, 3, 2, 1, 4, 5],
    'quantite' : [3, 1, 5, 1, 1, 8],
    'data' : ['2024-11-10', '2024-11-10', '2024-11-10', '2024-11-11', '2024-11-11', '2024-11-11']
}
commandes_df = pd.DataFrame(commandes_data)
print("commandes creees:")
commandes_df
```

```python
#algorithme de recommandation

class SystemeRecommandation:
    def __init__(self, plats_df, commandes_df):
        self.plats_df = plats_df
        self.commandes_df = commandes_df
        self.vectorizer = TfidfVectorizer()
    def preparer_donnees(self):
        """prepare les donnnees pour l'analyse"""
        
        #combiner les caraceristiques
        self.plats_df['caracteristiques'] = (
          self.plats_df['categorie'] + ' ' + self.plats_df['ingredients']
        )
        #vectoriser
        self.vectors = self.vectorizer.fit_transform(
            self.plats_df['caracteristiques']
        )
        print("donnees preparees")
    def recommander_par_similarite(self, nom_plat, n_recommandations=3):
        """recommande les plat similaires"""
        
        # Trouver la catégorie du plat demandé
        categorie = self.plats_df[self.plats_df['nom'] == nom_plat]['categorie'].iloc[0]
        # Retourner les autres plats de la même catégorie
        similaires = self.plats_df[
            (self.plats_df['categorie'] == categorie) & 
            (self.plats_df['nom'] != nom_plat)
        ]
        return similaires
 
    def recommander_populaires(self):
        """recommande les plats les plus populaires"""
        
        popularite = self.commandes_df.groupby('id_plat')['quantite'].sum()
        
        top_3_ids = popularite.nlargest(3).index
        return self.plats_df.loc[self.plats_df['id'].isin(top_3_ids)]
        
    def plats_par_categorie(self, categorie):
        """Retourne tous les plats d'une catégorie"""
        return self.plats_df[self.plats_df['categorie'] == categorie]

#initialiser le systeme
systeme = SystemeRecommandation(plats_df, commandes_df)
systeme.preparer_donnees()
        
        
        
        
        
        
```

Tests du systeme


```python
print("Test du systeme de recommandation")
print("=" * 40)

#test 1 recommandation par similarite
print("reccommandations pour 'poulet DG' :")
recommandations = systeme.recommander_par_similarite("poulet DG")
print("Plats similaires recommandes: ")
if not recommandations.empty:
  for _, plat in recommandations.iterrows():
    print(f" {plat['nom']} - {plat['prix']} FCFA ")
else:
    print("Aucune recommandation trouvee")
#test 2 plats populaires
print("Plats les plus populaires:")
populaires = systeme.recommander_populaires()
if not populaires.empty:
  for _, plat in populaires.iterrows():
    print(f" {plat['nom']} - {plat['prix']} FCFA")
else:
    print("Aucun plat populaire trouve")  

# 2. Plats par catégorie
print(" PLATS VIANDE :")
viandes = systeme.plats_par_categorie("viande")
for _, plat in viandes.iterrows():
    print(f"• {plat['nom']} - {plat['prix']} FCFA")

print(" PLATS LÉGUMES :")
legumes = systeme.plats_par_categorie("legumes")
for _, plat in legumes.iterrows():
    print(f"• {plat['nom']} - {plat['prix']} FCFA")
```
